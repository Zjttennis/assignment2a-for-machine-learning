# Reflection

## Definition of reinforcement learning

Reinforcement learning is a framework for solving control tasks (also called decision problems) by building agents that learn from the environment by interacting with it through trial and error and receiving rewards (positive or negative) as unique feedback.

## Basic knowledge

- reinforcement learning is a circle learing, it's process is a loop:state, action, reward and next state. The final target is maximize rewards.

### State

- State: complete description of state of the world.
- Compared to state, there is also the concept of observation: partial description of the state of the world.

### Action

- Action is limited: discrete space.
- Action is infinite: continuous space.

## Task

- Episodic tasks.
- Continuing tasks.  no terminal states.

## Gamma rate

### Basic knowledge of Gamma

- Gamma rate is a super parameter (discount factor), it is usually used to balance the weights of short-term rewards and long-term rewards.
- Gamma value is a number between 0 and 1, which is used for discount future rewards when accumlate rewards.
-  there is a equation for calculating cumulative rewards.

### Role of Gamma 

- Gamma approaches 1: emphasizes long-term rewards, and the model will pay more attention to future gains.
- Gamma approaches 0: emphasizes immediate rewards, and the model only pays attention to current gains and almost ignores future rewards.

### Question about Gamma

- Gamma is used to balance current benefits and future benefits. During the training process, the agent cannot know what will happen in the future. So when calculating the R(t) value, how should the prediction part be calculated?
- use value functions
- Fortunately, in unit 2, I learnt 2 value functions, state value function and action value function

## 2 main ways for RL problems

### Policy-based
- Train the policy directly to learn which action to take given a state.
    - deterministic plicy
    - stochastic policy

### Value based
- Train a value function to learn which state is more valuable and use this value function to take the action that leads to it.
    - state value function
    - action value function

- In value based function, I learnt an Epsilon-Greedy Policy, also there is a Greedy Policy. For Greedy Policy, it always chooses the currently known best action, that is, in each state, it chooses the action that maximizes the immediate reward. But for Epsilon-Greedy Policy. it is an extended strategy with randomness. It selects greedy actions most of the time (probability:1-Epsilon), but selects random actions a small part of the time (probability: Epsilon) for exploration.

- State value function
    - Calculate the value of state St
- Action value function
    - Calculate the value of (State, action) pair

## There is a relationship between policy and value: an optimal value function leads to have an optimal policy

## Bellman function

- The Bellman equation describes the state value function or action value using recursion.
     - The value of the state V(St) is equal to the immediate reward R(t+1) + the discounted (gamma) value of next state(V(St+1)) if agent starts at state S (St)

## Mente Carlo

- Learning is based on a complete episode.
- Each update requires waiting for the episode end, calculating the actual total reward (Gt) starting from the current state, and then using it to update the value function V(St).

## TD

- Based on single-step updates, there is no need to wait for the end of the episode.
- Use "current reward + estimated value of next state" as the target value for update.

## Q-learning

- Q-learning uses Q function to estimate the expected reward for each state-action pair. By updating the Q-value, the agent gradually learns the best action to take in each state.
- When I was learning Q-learning, I also learnt 2 new concetp, off-policy and on policy.
    - On-policy algorithms require that the agent uses the strategy for experience when learning,the strategy for acting is the same as the updating strategy.
    - Off-policy, using a different policy for acting strategy and updating strategy.

- There is a basic Q-learning process
    - First: Initialize Q table, set all elements as 0.
    - Second: Choose an action using Epsilon Greedy Strategy.
    - Third: Perform action At, get Rt+1 and St+1.
    - Update Q pair(St,At).
    - Loop from second step until episode end.

### Also, optimal Q-function = optimal policy
